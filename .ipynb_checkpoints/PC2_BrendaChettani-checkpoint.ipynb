```python
# Con skiprows estoy saltando las primeras 4 filas y el header seria la 5ta fila
# Con nrows estoy leyendo el numero de filas que indico ahi (225) desde la fila que tengo (5)
import pandas as pd
tabla1 = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRpSTUr2WXniw0ysnXT2blO9UELpYYoISFHjXVtoDdaFEPXoykVbpX7FXMKCLa9cFyZvpnWDspn-iQe/pub?gid=65778500&single=true&output=csv', skiprows=4, nrows=225)
tabla1
```

```python
# Con drop voy a borrar las columnas vacias o con datos no utiles
tabla1= tabla1.drop(tabla1.columns[[3, 5, 7, 9, 11, 13]], axis=1)
tabla1
```

```python
# Limpio los espacios en cada celda
tabla1=tabla1.apply(lambda x: x.str.strip())
```

```python
# Remplazo las comas por punto
tabla1=tabla1.replace(',','.',regex=True)
tabla1
```

```python
# Remplazo .. o — por NaN
import numpy as np
tabla1.replace({'..': pd.NA, '—': pd.NA}, inplace=True)
tabla1
```

```python
#Quise ver el nombre de esa columna porque en la siguiente no me cambiaba el nombre
tabla1.columns[2]
```

```python
# Con rename estoy cambiando el nombre de las columnas
# Con drop estoy eliminando la fila 0 y 1 ya que no tiene datos utiles
tabla1 = tabla1.rename(columns={'Unnamed: 0':'HDI rank', 'Unnamed: 1':'Country', 'Human Development Index (HDI) ':'Human Development Index (HDI) (2021)', 'Life expectancy at birth': 'Life expectancy at birth (2021)', 'Expected years of schooling':'Expected years of schooling (2021)', 'Mean years of schooling':'Mean years of schooling (2021)', 'Gross national income (GNI) per capita':'Gross national income (GNI) per capita (2021)', 'GNI per capita rank minus HDI rank':'GNI per capita rank minus HDI rank (2021)', 'HDI rank':'HDI rank (2020)'})
tabla1 = tabla1.drop(tabla1.index[[0,1]])
tabla1
```

```python
# Convierto todas las columnas, menos Country, a float
# Veo el tipo de dato
columnas_a_convertir=tabla1.columns.drop(['Country'])
tabla1[columnas_a_convertir] = tabla1[columnas_a_convertir].apply(pd.to_numeric, errors='coerce')
tabla1.info()
```

```python
# Verifico que no se repitan los paises (como vemos hay algunos espacios con NaN en la columna de Coountry y por eso sale falso)
len(tabla1.Country)==len(set(tabla1.Country))
```

```python
# Veo que filas tienen NaN
tabla1[tabla1.isna().any(axis=1)]
```

```python
# Borro las filas en donde todas las columnas son NaN
tabla1= tabla1.dropna(how='all')
# Veo que filas tienen NaN ahora
tabla1[tabla1.isna().any(axis=1)]
```

```python
# Veo que las filas (211) Regions, (203) Human development groups y (197) OTHER COUNTRIES OR TERRITORIES estan demas pero las mantengo porque son la separacion de los otros incisos. Lo necesito y no se me ocurre otra idea de agruparlo
# Pero si borrare las filas (164) LOW HUMAN DEVELOPMENT, (119) MEDIUM HUMAN DEVELOPMENT, (69) HIGH HUMAN DEVELOPMENT, (2) VERY HIGH HUMAN DEVELOPMENT
# Verifico
tabla1=tabla1.drop([2, 69, 119, 164])
tabla1[tabla1.isna().any(axis=1)]
```

```python
# Por otro lado, veo que la (164) LOW HUMAN DEVELOPMENT, (119) MEDIUM HUMAN DEVELOPMENT, (69) HIGH HUMAN DEVELOPMENT, (2) VERY HIGH HUMAN DEVELOPMENT, para estos casos, como se repiten en gran parte del cuadro, creare una nueva columna 'Human development groups'
valores = ['VERY HIGH HUMAN DEVELOPMENT'] * 66 + ['HIGH HUMAN DEVELOPMENT'] * 49 + ['MEDIUM HUMAN DEVELOPMENT'] * 44 + ['LOW HUMAN DEVELOPMENT'] * 32 + ['NAN'] * 22
valores_repetidos = (valores * ((len(tabla1) // len(tabla1)) + 1))[:len(tabla1)]
tabla1['Human development groups']=valores_repetidos
tabla1
```
